```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

def stretch_histogram(image_path):
    # Ouvrir l'image et convertir en niveaux de gris
    im = Image.open(image_path).convert('L')  # Convertir en niveaux de gris

    # Convertir l'image en tableau numpy
    tab = np.array(im)

    # Appliquer l'étirement de l'histogramme
    stretched_tab = cv2.normalize(tab, None, 0, 255, cv2.NORM_MINMAX)

    # Afficher l'image transformée
    plt.imshow(stretched_tab, cmap='gray')
    plt.axis('off')  # Masquer les axes
    plt.title('Image avec étirement de l\'histogramme')
    plt.show()

    # Calculer et afficher l'histogramme de l'image transformée
    hist_stretched = cv2.calcHist([stretched_tab], [0], None, [256], [0, 256])

    # Afficher l'histogramme de l'image transformée
    plt.plot(hist_stretched, color='blue')
    plt.title('Histogramme après étirement')
    plt.xlabel('Valeur des pixels')
    plt.ylabel('Fréquence')
    plt.xlim([0, 256])
    plt.show()

# Utiliser la fonction prédéfinie
stretch_histogram('Crayfish_NG.jpg')
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

def equalize_histogram(image_path):
    # Ouvrir l'image et convertir en niveaux de gris
    im = Image.open(image_path).convert('L')

    # Convertir l'image en tableau numpy
    tab = np.array(im)

    # Appliquer l'égalisation de l'histogramme
    equalized_tab = cv2.equalizeHist(tab)

    # Afficher l'image transformée
    plt.imshow(equalized_tab, cmap='gray')
    plt.axis('off')
    plt.title('Image avec égalisation d\'histogramme')
    plt.show()

    # Calculer et afficher l'histogramme de l'image transformée
    hist_equalized = cv2.calcHist([equalized_tab], [0], None, [256], [0, 256])
    plt.plot(hist_equalized, color='blue')
    plt.title('Histogramme après égalisation')
    plt.xlabel('Valeur des pixels')
    plt.ylabel('Fréquence')
    plt.xlim([0, 256])
    plt.show()

# Utiliser la fonction prédéfinie
equalize_histogram('Crayfish_NG.jpg')
```
